# IMPORTING LIBRARIES NUMPY,MATH,MATPLOTLIB, PANDAS

```python
import numpy as np
import math
from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
```

# READING THE CSV FILE INTO THE DATAFRAME

```python
df = pd.read_csv('C:/Users/sarang.kulkarni/Desktop/DATA MINING/DATA TRANSFORMATION/cereal.csv')    #paste file path and name here
```

# PLOTTING HISTOGRAM

```python
plt.hist(df.loc[:,"calories"])  #relplace "calories" by the column(feature) u want
plt.title("Histogram for feature 1: Calories ")   #relplace "Pregnancies" by the column(feature) u want
plt.xlabel("Calories")      #relplace "Pregnancies" by the column(feature) u want
plt.ylabel("Count")

plt.show()
```

# Calculate Mean

```python
cal_mean=df.loc[:,"calories"].mean()
```

# Calculate Standard Deviation

```python
cal_std=df.loc[:,"calories"].std()
```

# Z Score = (X - Mean)/ (Stand. Deviation. )

```python
half_df_cal= df.loc[:,"calories"]-cal_mean
```

```python
Normal_df_cal=half_df_cal/cal_std
```

# PLotting the Distribution/Histogram of Normalised Column attribute

```python
plt.hist(Normal_df_cal)  #relplace "calories" by the column(feature) u want
plt.title("Histogram for feature 1: Calories ")   #relplace "Pregnancies" by the column(feature) u want
plt.xlabel("Calories (Z score)")      #relplace "Pregnancies" by the column(feature) u want
plt.ylabel("Count")

plt.show()
```
