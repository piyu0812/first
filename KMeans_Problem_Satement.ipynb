Perform clustering on Age and spending score


```python
df1=pd.read_csv('Mall_Customers.csv')
df1
```

```python
a_s=df1[['Age','Spending Score (1-100)']]
a_s
```

```python
wcc=[]
for i in range(1,10):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(a_s)
  wcc.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10),wcc)
```

Here we find that the efficient clusters is 4

```python
x=a_s.iloc[:,:].values
kmeans=KMeans(n_clusters=4, init='k-means++',random_state=42)
y=kmeans.fit_predict(x)
y
centroid=kmeans.cluster_centers_
centroid

```

```python
plt.scatter(x[y==0,0],x[y==0,1],color='red')
plt.scatter(x[y==1,0],x[y==1,1],color='green')
plt.scatter(x[y==2,0],x[y==2,1],color='blue')
plt.scatter(x[y==3,0],x[y==3,1],color='black')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.scatter(centroid[:, 0], centroid[:, 1], marker='X', s=200, c='cyan')
```
